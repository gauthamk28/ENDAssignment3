#### 1.Write a function using only list filter lambda that can tell whether a number is a Fibonacci number or not. You can use a pre-calculated list/dict to store fab numbers till 10000

```
import math 

fibList = []

def isPerfectSquare(x): 
    s = int(math.sqrt(x)) 
    return s*s == x 
  
def isFibonacci(n):
  return isPerfectSquare(5*n*n + 4) or isPerfectSquare(5*n*n - 4) 
     

for i in range(1,10000): 
     if (isFibonacci(i) == True): 
         fibList.append(i)
     else: 
         pass

inp_value = input("Enter a number between 1 to 10000:")

print("Fibonacci numbers between 1 to 10000 is : ",fibList)


result = lambda x : "You have entered a Fibonacci number" if int(x) in fibList else "The number you have entered is not a fibonacci number"
print(result(inp_value))
```

#### 2.Using list comprehension (and zip / lambda / etc if required) write five different expressions that: PTS:100

1.    add 2 iterables a and b such that a is even and b is odd
2.    strips every vowel from a string provided (tsai>>t s)
3.    acts like a ReLU function for a 1D array
4.    acts like a sigmoid function for a 1D array
5.    takes a small character string and shifts all characters by 5 (handle boundary conditions) tsai>>yxfn



```
#1. add 2 iterables a and b such that a is even and b is odd
import random


checkListA = random.sample(range(5, 170), 20)
checkListB = random.sample(range(180, 369), 20)

print("List A : ",checkListA)
print("List B : ",checkListB)

eAndOddList = list(filter(lambda x: (x % 2 == 0), checkListA)) + list(filter(lambda y: (y % 2 != 0), checkListB))

result = sum(list(filter(lambda x: (x % 2 == 0), checkListA)) + list(filter(lambda y: (y % 2 != 0), checkListB)))


print("Even numbers from List A followed by odd numbers from List B",eAndOddList)

print("Sum of even and odd numbers from the list is: ",result)
```

```
#2. strips every vowel from a string provided (tsai>>t s)
inp_string = "tsai"

#with list comprehension
result = ' '.join(list(filter(lambda x: x not in "aeiou", inp_string)))

print(result)
```

```
#3. acts like a ReLU function for a 1D array


inpList = []

for i in range(0,10):
  rn = round(random.uniform(-1.0, 1.0),4)
  inpList.append(rn)

print("Input 1D array is : ",inpList)

result = [(lambda x: x if x > 0 else 0)(x) for x in inpList]

print("ReLU result : ",result)
```

```
#4. acts like a sigmoid function for a 1D array

import math

inpList = []

for i in range(0,10):
  rn = round(random.uniform(-1.0, 1.0),4)
  inpList.append(rn)

print("Input 1D array is : ",inpList)

result = [(lambda x: 1 / (1 + math.exp(-x)))(x) for x in inpList]

print("Sigmoid result : ",result)
```

#### 4.Using reduce functions: PTS:100

1.    add only even numbers in a list
2.    find the biggest character in a string (printable ascii characters)
3.    adds every 3rd number in a list



```
#1. add only even numbers in a list
from functools import reduce

sample_list = random.sample(range(5, 170), 20)

# add only even numbers in a list
evn_nos = list(filter(lambda x: x % 2 == 0, sample_list))
print('the list is ' , sample_list)
print ("even numbers from the list:" + str(evn_nos))
print("sum of even numbers")
reduce(lambda a, b: a + b, evn_nos)
```

```
#2. find the biggest character in a string (printable ascii characters)
big_string = "this is the third assignment in END yippeee $"
print('string is: ' ,big_string)
print('biggest character is: ')
reduce(lambda a, b: max(a, b), list(big_string))
```

```
#3. adds every 3rd number in a list

n = 3
lst = random.sample(range(5, 170), 20)
lst_3rd = lst[(n-1)::n] 
print('the list is ',lst)
print("every 3rd element from the list:" ,lst_3rd)
print("sum of all 3rd elements")
reduce(lambda a, b: a + b, lst_3rd)
```
